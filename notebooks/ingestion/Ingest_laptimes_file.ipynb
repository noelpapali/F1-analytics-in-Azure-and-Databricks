### ingest laptimes file

##### Read the multiple csv files using the spark dataframe reader

```python
dbutils.widgets.text("p_data_source", "")
v_data_source = dbutils.widgets.get("p_data_source")
```

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType
```

```python
laptimes_schema = StructType([
    StructField("raceId", IntegerType(), False),
    StructField("driverId", IntegerType(), True),
    StructField("lap", IntegerType(), True),
    StructField("position", IntegerType(), True),
    StructField("time", StringType(), True),
    StructField("milliseconds", IntegerType(), True)
])
                               
```

```python
laptimes_df = spark.read \
.schema(laptimes_schema) \
.csv("/mnt/f1datalake2025/raw/lap_times")
```

```python
display(laptimes_df)
```

```python
laptimes_df.count()
```

#####add ingestion date and  rename


```python
from pyspark.sql.functions import current_timestamp
```

```python
laptimes_final_df = laptimes_df.withColumnRenamed("raceId", "race_id") \
.withColumnRenamed("driverId", "driver_id") \
.withColumn("ingestion_date", current_timestamp())
```

######
write the file as parquet to cleaned and processed container

```python
laptimes_final_df.write.mode("overwrite").parquet("dbfs:/mnt/f1datalake2025/cleaned-and-processed/laptimes")
```
